```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/IPL.csv", on_bad_lines='skip', delimiter=',')
df
```

```python
print(df.isnull().sum())
```

Dropping unnecessary columns

```python
# Display the existing columns in your DataFrame
df.drop(columns=['Unnamed: 0', 'next_batter'], inplace=True, errors='ignore')
#                                 ^ Added space
# errors='ignore' will skip any non-existent columns to avoid the KeyError.
```

```python
print(df.columns)
```

```python
# Fill missing values in 'event_name' with 'Unknown'
df['event_name'].fillna('Unknown', inplace=True)

# Fill missing values in numerical columns with 0
df['curr_batter_runs'].fillna(0, inplace=True)
df['curr_batter_balls'].fillna(0, inplace=True)
df['bowler_wicket'].fillna(0, inplace=True)
```

```python
df.drop(columns=['batting_partners'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
# Convert 'date' to datetime, handling errors
df['date'] = pd.to_datetime(df['date'], errors='coerce', format='%Y-%m-%d')

# Convert categorical columns to 'category' type
df['match_type'] = df['match_type'].astype('category')
df['event_name'] = df['event_name'].astype('category')
```

```python
# Convert 'runs_batter' to numeric, handling errors
df['runs_batter'] = pd.to_numeric(df['runs_batter'], errors='coerce')

# Check for negative values in 'runs_batter'
print(df[df['runs_batter'] < 0])

# Replace negative values with 0
df['runs_batter'] = df['runs_batter'].clip(lower=0)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
df.dropna(subset=['date'], inplace=True)
# Fill missing 'innings' with the mode
df['innings'].fillna(df['innings'].mode()[0], inplace=True)
# Fill missing 'batting_team_players' with 'Unknown'
df['batting_team_players'].fillna('Unknown', inplace=True)
# Fill missing 'bowling_team_players' with 'Unknown'
df['bowling_team_players'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.to_csv('cleaned_ipl_dataset.csv', index=False)
```
